```python
import requests
import selenium
import numpy as np
import pandas as pd
from PIL import Image
from io import BytesIO
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.edge.service import Service as EdgeService
```

```python
class jobs_google(object):
    def web_scrapping(self):
        '''
        This function scrape google job site and extract informations like company name , job title , location ,
        posted date , salary and company logos
    
        Parameters
        ----------
        url is given inside the function itself 

        Returns
        -------
        df - Dataframe containing all the above mention details
        csv file - csv file containg all the data present in the dataframe
   
        '''
        
        
        
        

        with requests.Session() as s:
            url = f"https://www.google.com/search?q=jobs+near+me"
            headers = {
            "referer":"referer: https://www.google.com/",
            "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36"
            }
            s.post(url, headers=headers)
            response = s.get(url, headers=headers)
            soup = BeautifulSoup(response.text, 'html.parser')
            url =soup.find_all("a",{"class":"esVihe"})
        with requests.Session() as s:
            url = url[0].get("href")
            headers = {
            "referer":"referer: https://www.google.com/",
            "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36"
            }
            s.post(url, headers=headers)
            response = s.get(url, headers=headers)
            soup = BeautifulSoup(response.text, 'html.parser')
            # extracting job title from web page
            jobs = soup.find_all("div",{"class":"BjJfJf PUpOsf"})
            joblist = []
            joblist = [val.text for val in jobs]
            # extracting location from web page
            location = soup.find_all("div",{"class":"Qk80Jf"})
            loc = ([val.text for val in location])
            locatn = [loc[i] for i in range(len(loc)) if i%2==0]
            # extracting company name from web page
            company_name = soup.find_all("div",{"class":"vNEEBe"})
            Company_names = []
            Company_names=[val.text for val in company_name]
            # extracting salary from the web page
            salary=[]
            sal = soup.find_all("div",{"class":"KKh3md"})
        for post in sal:
            date = post.find("span",{"class":"LL4CDc"}).text
            if date.startswith("₹")==True:
                salary.append(date)
            else:
                salary.append("Not Available")
            # extracting posted date from web page
            pos=[]
            posted = soup.find_all("div",{"class":"KKh3md"})
        for post in posted:
            posted_date = post.find("span",{"class":"LL4CDc"}).text
            if posted_date.endswith("ago")==True:
                pos.append(posted_date)
            else:
                pos.append("Not Available")
        # extracting  company logo from web page
        url = f"https://www.google.com/search?q=jobs+near+me"
        driver = webdriver.Edge(service=EdgeService(EdgeChromiumDriverManager().install()))
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,features = "html.parser")
        web_url = soup.find("a",{"class":"esVihe"})
        main_url = str(web_url.get("href"))
        driver.get(main_url)
        soup_sample = BeautifulSoup(driver.page_source, "html.parser" )
        logo_path = soup_sample.find_all("img",{"class":"YQ4gaf zr758c"})
        img = []
        for job in logo_path:
            images = job.get("src")
            response = requests.get(images)
            img.append(np.array(Image.open(BytesIO(response.content))))

            
            
            
        # creating dictonary
        data = {"COMPANY NAME":Company_names,"JOB ": joblist,"LOCATION":locatn,"POSTED DATE":pos,"SALARY":salary,"COMPANY LOGO":img}
        #Creating DataFrame
        df = pd.DataFrame(data)
        # creating the  csv file from dataframe
        df.to_csv('dsa_internship_jobsearch7.csv')
        return df
```
